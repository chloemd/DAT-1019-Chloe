There are many null values in this dataset. I will examine the dataset column by column to see what reasonable values for the different nulls should be.

```python
def print_nulls(df):
    for column in df.columns:
        print(f"{column} Nulls: {df[df[column].isnull()].shape[0]}")
        print(f"{column} Valids: {df[df[column].notnull()].shape[0]}")

def nulls_to_df(df):
    """Returns a dataframe with a count of null and valid entries for each column in the
    original dataset."""
    d = dict()
    for column in df.columns:
        l = []
        l.append(df[df[column].isnull()].shape[0])
        l.append(df[df[column].notnull()].shape[0])
        d[column] = l
    
    new_df = pd.DataFrame.from_dict(d, orient='index', columns=['Nulls', 'Valids'])
    return new_df
```

```python
print_nulls(df)
```

```python
nulls = nulls_to_df(df)
nulls.sort_values(by='Nulls', ascending=False)
```

**Range (FT2)** is trickier. The column denotes the range of vehicles using their second fuel type (FT2 = fuel type 2). The integer-like values of this column are an average of the City Range (FT2) and Highway Range (FT2) columns. However, inspecting the datatypes of this dataframe, I can see that there are also some string values in this column. I need to find these, make sense of them, and, if necessary, convert them to integers. 

```python
df[df['Range (FT2)'].apply(type) == str]
```

```python
df[df['Range (FT2)'].apply(lambda x: len(str(x)) > 3)]
```

Additional research online revealed that the "slash" notation (e.g. 230/270/270) in the Range (FT2) column denotes differences between pickup trucks with "regular cab short wheelbase, regular cab long wheelbase, and super cab models, respectively". I found this information in a report on [FuelEconomy.gov](https://www.fueleconomy.gov/FEG/pdfs/guides/FEG2000_Part1.pdf).

To simplify this column, I think it's reasonable to take the average of the values and use that single value instead. First, though, I will replace null values with 0.

```python
df['Range (FT2)'].fillna(0, inplace=True)
```

```python
def convert_rangeft(series):
    """Returns a version of the Range (FT#) column with only integers and 
    uses the mean for rows with multiple values listed."""
    l = []
    for i in series:
        try:
            l.append(int(i))
        except:
            if '/' in i:
                i = i.split('/')
            elif '-' in i:
                i = i.split('-')
            i = [int(num) for num in i]
            avg = round(sum(i)/len(i))
            l.append(avg)
    return pd.Series(l)
            
                
```

```python
df['Range (FT2) Clean'] = convert_rangeft(df['Range (FT2)'])
```

```python
df[df['Range (FT2) Clean'] > 0]
```

According to FuelEconomy.gov, the **Start-Stop Technology** field is left blank for older vehicles. The latest model year for vehicles with null values is 2012. However, according to the [Wikipedia page](https://en.wikipedia.org/wiki/Start-stop_system#History) for this technology, there were vehicles using a version of this technology as early as 1974, so I can't just assume that all older models lacked this feature. Given that the original data treats "no" and "unknown" as categorically different, I will replace the nulls with "Unknown".

```python
df['Start Stop Technology'].fillna('Unknown', inplace=True)
nulls = nulls_to_df(df)
nulls.sort_values(by='Nulls', ascending=False)
```

```python
man_codes = df[df['Manufacturer Code'].notnull()][['Make','Manufacturer Code']].groupby(by='Make').describe()
man_codes.sort_values(by=('Manufacturer Code', 'unique'), ascending=False)
```

Most makes correspond with a single **Manufacturer Code**. For makes with more than one, it makes sense to use the more frequent value. I can do this by mapping. If the list of null manufacturers codes contains makes that are not in this list, I will use "Unknown" instead.

```python
top_codes = dict(man_codes[('Manufacturer Code', 'top')])
top_codes
```

```python
df['Manufacturer Code Clean'] = np.where(df['Manufacturer Code'].isnull(), df['Make'].map(top_codes) ,df['Manufacturer Code'])
```

```python
df[df['Manufacturer Code Clean'].isnull()]
```

This reduced the number of nulls from over 30,000 to around 3,000. The remaining 3,000 nulls correspond to vehicles whose make was not in the top_codes list. I will replace them with "Unknown" then drop the old Manufacturer Code column:

```python
df['Manufacturer Code Clean'].fillna('Unknown', inplace=True)
df = df.drop('Manufacturer Code', axis=1)
nulls = nulls_to_df(df)
nulls.sort_values(by='Nulls', ascending=False)
```
