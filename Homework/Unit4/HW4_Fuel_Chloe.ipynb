There are many null values in this dataset. I will examine the dataset column by column to see what reasonable values for the different nulls should be.

```python
def print_nulls(df):
    for column in df.columns:
        print(f"{column} Nulls: {df[df[column].isnull()].shape[0]}")
        print(f"{column} Valids: {df[df[column].notnull()].shape[0]}")

def nulls_to_df(df):
    """Returns a dataframe with a count of null and valid entries for each column in the
    original dataset."""
    d = dict()
    for column in df.columns:
        l = []
        l.append(df[df[column].isnull()].shape[0])
        l.append(df[df[column].notnull()].shape[0])
        d[column] = l
    
    new_df = pd.DataFrame.from_dict(d, orient='index', columns=['Nulls', 'Valids'])
    return new_df
```

```python
nulls = nulls_to_df(df)
nulls.sort_values(by='Nulls', ascending=False)
```

With 38,067 null values and only 46 valid values, **Alternate Charger** is in the worst shape, so I'll start with that. Based on the values in the column and on information found on the [Department of Energy's website](https://www.fueleconomy.gov/feg/ws/), this column seems to describe alternate chargers for electric vehicles. This does not apply to the vast majority of vehicles, so a reasonable value to substitute is "None".

```python
df['Alternate Charger'].unique()
```

```python
df['Alternate Charger'].fillna('None', inplace=True)
df['Alternate Charger'].unique()
```

```python
nulls = nulls_to_df(df)
nulls.sort_values(by='Nulls', ascending=False)
```

**Range (FT2)** is trickier. The column denotes the range of vehicles using their second fuel type (FT2 = fuel type 2). The integer-like values of this column are an average of the City Range (FT2) and Highway Range (FT2) columns. However, inspecting the datatypes of this dataframe, I can see that there are also some string values in this column. I need to find these, make sense of them, and, if necessary, convert them to integers. 

```python
df[df['Range (FT2)'].apply(type) == str]
```

```python
df[df['Range (FT2)'].apply(lambda x: len(str(x)) > 3)]
```

Additional research online revealed that the "slash" notation (e.g. 230/270/270) in the Range (FT2) column denotes differences between pickup trucks with "regular cab short wheelbase, regular cab long wheelbase, and super cab models, respectively". I found this information in a report on [FuelEconomy.gov](https://www.fueleconomy.gov/FEG/pdfs/guides/FEG2000_Part1.pdf).

To simplify this column, I think it's reasonable to take the average of the values and use that single value instead. First, though, I will replace null values with 0.

```python
df['Range (FT2)'].fillna(0, inplace=True)
```

```python
nulls = nulls_to_df(df)
nulls.sort_values(by='Nulls', ascending=False)
```
